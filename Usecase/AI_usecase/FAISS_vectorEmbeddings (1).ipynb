```
# Create dataset (tagged into category as 'Health', 'Fashion', 'Technology', 'Emotion')
data = [
    ['I like to eating healthy vegetables and fruits', 'Health'],
    ['Chicken is a good source of protein', 'Health'],
    ['Exercise daily for fitment', 'Health'],
    ['I love blue color clothes', 'Fashion'],
    ['Long shoes are always classic', 'Fashion'],
    ['Black shirt is good choice', 'Fashion'],
    ['electric Vehicles are the FUTURE', 'Technology'],
    ['People nowadays usage aaple gadhets more', 'Technology'],
    ['NvDiA is coming now with faster GPU', 'Technology'],
    ['Im feeling happy today', 'Emotion'],
    ['happy, sad are my moods when I saw you', 'Emotion'],
    ['Always girls are happy when they see rich men', 'Emotion']
]

import pandas as pd
df = pd.DataFrame(data, columns=['Text', 'Category'])
df
```

```
# Data preprocessign - cleaning dataset using 'Spacy' module
import spacy
nlp = spacy.load('en_core_web_sm')
stop_words = nlp.Defaults.stop_words

def preprocess(text):
  doc = nlp(str(text))
  preprocessed_text = []
  for token in doc:
    if token.is_punct or token.like_num or token in stop_words or token.is_space:
      continue
    preprocessed_text.append(token.lemma_.lower().strip())
  return ' '.join(preprocessed_text)

df['Processed_text'] = df['Text'].apply(preprocess)
df
```

```
# Load pre-trained sentence embedding model
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')

df['Embeddings'] = df['Processed_text'].apply(model.encode)
vector = model.encode(df['Processed_text'])
df
```

```
import faiss                      # Imports FAISS, which is optimized for fast similarity search.
dim = vector.shape[1]             # Extracts the embedding dimension (384 for all-MiniLM-L6-v2).
print(dim)                        # Maps sentences to 384-dimensional dense vector embeddings
index = faiss.IndexFlatL2(dim)    # Initializes a flat L2 (Euclidean distance) index.
index.add(vector)                 # Adds your sentence embeddings to the index.
```

```
# User input query
search_query = 'Google notebooks are how much powerfull?'

test_pre = preprocess(search_query) # Pre-processing text
encode_pre = model.encode(test_pre) # Converting to vector embeddings
encode_pre.shape

# FAISS expects 2D array, we are converting encode_pre to a 2D array
import numpy as np
svec = np.array(encode_pre).reshape(1,-1)
```

```
# Get euclidean distance and index of the 2 nearest neighbours
distance,pos = index.search(svec, k=2)
print(distance, pos)
```

```
# Check - user query is similar to which category
df.loc[pos[0], ['Text', 'Category']]
```
